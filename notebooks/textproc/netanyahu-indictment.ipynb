```python
# text analysis of netanyahu's indictment 

path = "file_03.01.2021.pdf"

import os
import textract
import nltk
```

```python
raw = textract.process(path).decode('utf8')
tokens = nltk.word_tokenize(raw)
porter = nltk.PorterStemmer()
text = nltk.Text([porter.stem(t) for t in tokens])
```

```python
text.concordance("שוחד")
```

```python
text.similar("נתניהו")
```

```python
text.common_contexts(["נתניהו", "אלוביץ"])
```

```python
text.dispersion_plot(["נתניהו", "אלוביץ", "מוזס", "פילבר", "שרה", "חפץ", "מילצ"])
```

```python
text.generate()
```

```python
# lexical richness
len(set(text)) / len(text)
```

```python
# counting, percentage
for word in ["נתניהו", "הנאשם", "אלוביץ"]:
    p = text.count(word) / len(text)
    print("{}: {:2.2%}".format(word, p))
```

```python
fdist = nltk.FreqDist(text)
print(fdist)
fdist.most_common(10)
```

```python
fdist.plot(50, cumulative=False)
```

```python
# words showing only once
fdist.hapaxes()
```

```python
# long words
V = set(text)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```
